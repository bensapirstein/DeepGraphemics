```python
import matplotlib.pyplot as plt
from utils.datasets import GraphemesDataset
# Example usage:
root_dir = "dataset_skia"
train_dataset = GraphemesDataset(root_dir, train=True, by_letter='aleph')
test_dataset = GraphemesDataset(root_dir, train=False, by_letter='aleph')
```

```python
# Accessing dataset samples
image, label = train_dataset[0]
print(f"Train Image shape: {image.shape}, Label: {label}")

image, label = test_dataset[0]
print(f"Test Image shape: {image.shape}, Label: {label}")
```

```python
# plot the first 10 images in the train dataset
fig, axes = plt.subplots(5, 5, figsize=(12, 8))
for i, ax in enumerate(axes.flatten()):
    image, label = train_dataset[i]
    ax.imshow(image[:, :, 0], cmap='gray')
    ax.set_title(train_dataset.classes[label])
    ax.axis('off')
plt.show()
```

```python
# plot the first 10 images in the test dataset
fig, axes = plt.subplots(5, 5, figsize=(12, 8))
for i, ax in enumerate(axes.flatten()):
    image, label = test_dataset[i]
    ax.imshow(image[:, :, 0], cmap='gray')
    ax.set_title(test_dataset.classes[label])
    ax.axis('off')
plt.show()
```

```python
# plot the distribution of labels in the train and test datasets and set the x-axis labels to the class names
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
for i, ax in enumerate(axes.flatten()):
    if i == 0:
        ax.hist([train_dataset[i][1] for i in range(len(train_dataset))])
        ax.set_title('Train')
    else:
        ax.hist([test_dataset[i][1] for i in range(len(test_dataset))])
        ax.set_title('Test')

axes[1].set_xticks(range(len(train_dataset.classes)))
axes[1].set_xticklabels(train_dataset.classes, rotation=90)
axes[0].set_xticks([])

plt.show()
```
